```python
from requests.adapters import HTTPAdapter
```

```python
from requests.exceptions import ConnectionError
```

```python
github_adapter = HTTPAdapter(max_retries=3)
session = requests.Session()
```

```python
session.mount("https://api.github.com", github_adapter)
```

```python
try:
    session.get("https://api.github.com")
except ConnectionError as ce:
    print(ce)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.moviechart.co.kr/rank/realtime/index/image"
movie_ranking = requests.get(url)

if movie_ranking.status_code == 200:
  print("영화 정보를 출력합니다.")
  soup = BeautifulSoup(movie_ranking.content)
  movie_list = soup.select(".movieBox-list .movie-title a")
  print(f"수집한 영화 수: {len(movie_list)}")
  for movie in movie_list:
    print(movie.text)
else:
  print("페이지에 연결할 수 없습니다.")
```

```python
import requests
from bs4 import BeautifulSoup

# 비동기 방식으로 페이지를 로드하는 사이트
url = "http://www.cine21.com/rank/boxoffice/domestic"
movie_ranking = requests.get(url)

if movie_ranking.status_code == 200:
  print("영화 정보를 출력합니다.")
  soup = BeautifulSoup(movie_ranking.content)
  boxoffice = soup.select_one("div#boxoffice_list_content")
  movie_list = boxoffice.select("li.boxoffice_li div.mov_name")
  print(f"수집한 영화 수: {len(movie_list)}")
  for movie in movie_list:
    print(movie.text)
else:
  print("페이지에 연결할 수 없습니다.")
```
