```python
from konlpy.corpus import kolaw
```

```python
data = kolaw.open("constitution.txt").read()
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
print(komoran.nouns("%r"%data[0:1000]))
```

```python
word_list = komoran.nouns("%r"%data[0:1000])
```

```python
text = ' '.join(word_list)
```

```python
text
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from wordcloud import WordCloud
```

```python
wordc = WordCloud()
wordc.generate(text)
```

```python
plt.imshow(wordc)
plt.show()
```

```python
wordc = WordCloud(background_color="white", max_words=20,
                  relative_scaling=0.2,
                  font_path="C:/Windows/Fonts/malgun.ttf")
wordc.generate(text)
plt.figure()
plt.imshow(wordc)
plt.axis("off")
```

```python
from sklearn.feature_extraction._stop_words import ENGLISH_STOP_WORDS
from wordcloud import STOPWORDS

stow_w = STOPWORDS | ENGLISH_STOP_WORDS | set(["대통령","국가"])
```

```python
wordcloud = WordCloud(background_color="white",
                      max_words=2000,
                      stopwords=stow_w,
                      relative_scaling=0.2,
                      font_path="C:/Windows/Fonts/malgun.ttf")
```

```python
wordcloud.generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
img = Image.open("snowman.png").convert("RGBA")
mask_ar = np.array(img)
```

```python
wordcloud = WordCloud(background_color="white",
                      max_words=2000,
                      font_path="C:/windows/Fonts/malgun.ttf",
                      mask=mask_ar, random_state=42)
```

```python
wordcloud.generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
#wordcloud.to_file("result1.png")
```

```python
import random
def grey_color(*args, **kwargs):
    return 'hsl(40, 100%%, %d%%)'% random.randint(50,100)
```

```python
wordcloud = WordCloud(background_color="black",
                      max_words=2000,
                      font_path="C:/windows/Fonts/malgun.ttf",
                      mask=mask_ar, random_state=42)
```

```python
wordcloud.generate(text)
wordcloud.recolor(color_func=grey_color)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
# wordcloud.to_file("result2.png")
```
