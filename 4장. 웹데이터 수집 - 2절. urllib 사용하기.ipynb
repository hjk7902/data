```python
import urllib.request as req
URL = "https://tomcat.apache.org"
header = {"User-Agent": "Mozilla/5.0 (iPhone)"}
request = req.Request(URL, headers=header) # request 객체를 생성함
print(request)
response = req.urlopen(URL) # 요청한 결과를 이용해서 response 객체를 생성함
print(response)
byte_data = response.read(100) # 00 바이트를 읽음, 기본값은 모든 데이터 읽음
print(byte_data)
```

```python
img_src = "https://tomcat.apache.org/res/images/tomcat.png"
req.urlretrieve(img_src, "tomcat.png")
```

```python
import urllib
parsed_url = urllib.parse.urlparse("https://tomcat.apache.org/?a=1&b=2&name=Kildong")
print(parsed_url)
```

```python
unparsed = urllib.parse.urlunparse(list(parsed_url))
print(unparsed)
```

```python
query_list = ["파이썬", "빅데이터", "인공지능"]
URL = "https://search.naver.com/search.naver?query="
```

```python
for query in query_list:
    new_URL = URL + urllib.parse.quote(query)
#     print(new_URL)
    data = req.urlopen(new_URL).read()
    print(data.decode())
```
